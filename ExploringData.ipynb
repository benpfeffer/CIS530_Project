### First: trying simple co-occurence of pizza type using cosine similarity

```python
simple = moDf[["order_id", "quantity", "pizza_type_id"]]
copyDf = simple.copy()
copyDf = copyDf[copyDf.quantity==1]
```

```python
copyDf = copyDf.drop_duplicates().reset_index(drop=True)
```

```python
itemitem = copyDf.pivot(index='pizza_type_id', columns='order_id', values='quantity').fillna(0)
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics.pairwise import pairwise_distances
```

```python
cositem = pd.DataFrame(pairwise_distances(itemitem,metric='euclidean'))
#cositem = pd.DataFrame(cosine_similarity(itemitem))
cositem.columns = itemitem.index
cositem.index = itemitem.index
```

```python
n=5
list(cositem[cositem > 0].stack().sort_values().head(n).index)
```

```python
cositem = pd.DataFrame(1 - pairwise_distances(itemitem,metric='cosine'))
#cositem = pd.DataFrame(cosine_similarity(itemitem))
cositem.columns = itemitem.index
cositem.index = itemitem.index
```

```python
n=10
list(cositem[cositem < 0.99].stack().sort_values(ascending=False).head(n).index)
```

```python
from sklearn.neighbors import NearestNeighbors
knn = NearestNeighbors(metric='cosine', algorithm='brute')
knn.fit(itemitem.values)
distances, indices = knn.kneighbors(itemitem.values, n_neighbors=5)
```

```python
distances.shape
```

```python
map_pizza = itemitem.index.to_list()
```

```python
print("Cosine similarity of best matches for each pizza:")
for i in range(len(indices)):
    curr = indices[i]
    dist = distances[i]
    print("_________________________________________________")
    print("Pizza:", map_pizza[curr[0]])
    print("Best matches:", [(map_pizza[j],round(1-k, 4)) for j, k in zip(curr[1:],dist[1:])])
```

### When people order multiple, rerun the code to recreate dataframe and develop a similarity matrix

```python
# Example, people ordered bbq_ckn and pepperoni together
```

```python
# Sum items and make new row that is a combo of two pizzas, remove old 2 pizzas, do same thing
```

```python
newRow = pd.DataFrame([np.where((itemitem.loc["bbq_ckn"] + itemitem.loc["pepperoni"] == 2)==True, 1, 0)])
newRow.columns = itemitem.columns
newRow.index = ["bbq_ckn-pepperoni"]
```

```python
orderTwo = pd.concat([itemitem, newRow], axis=0)
orderTwo = orderTwo.drop(["bbq_ckn", "pepperoni"])
```

```python
orderTwo.head()
```

```python
from sklearn.neighbors import NearestNeighbors
knn = NearestNeighbors(metric='cosine', algorithm='brute')
knn.fit(orderTwo.values)
distances2, indices2 = knn.kneighbors(orderTwo.values, n_neighbors=5)
```

```python
map_pizza2 = orderTwo.index.to_list()
```

```python
print("Cosine similarity of best matches for each pizza:")
for i in range(len(indices2)):
    curr = indices2[i]
    dist = distances2[i]
    print("_________________________________________________")
    print("Pizza:", map_pizza2[curr[0]])
    print("Best matches:", [(map_pizza2[j],round(1-k, 4)) for j, k in zip(curr[1:],dist[1:])])
```
